# Import necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load the dataset

```python
file_path = "IT_System_Analysis.xlsx"  # Update the file path if needed
df = pd.read_excel(file_path)
```

# Display the first few rows

```python
print("First 5 rows of the dataset:")
display(df.head())
```

# Basic information about the dataset

```python
print("\nDataset Information:")
df.info()
```

# Summary statistics

```python
print("\nSummary Statistics:")
display(df.describe())
```

# Check for missing values

```python
print("\nMissing Values:")
print(df.isnull().sum())
```

# Select only numeric columns

```python
numeric_df = df.select_dtypes(include=['number'])
```

# Correlation heatmap

```python
plt.figure(figsize=(12, 6))
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

# CPU Usage Distribution

```python
plt.figure(figsize=(8, 5))
sns.histplot(df["CPU Usage (%)"], bins=20, kde=True, color="blue")
plt.title("CPU Usage Distribution")
plt.xlabel("CPU Usage (%)")
plt.ylabel("Frequency")
plt.show()
```

# System Uptime vs. Server Downtime

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df["System Uptime (Hours)"], y=df["Server Downtime (Hours)"], color="red")
plt.title("System Uptime vs. Server Downtime")
plt.xlabel("System Uptime (Hours)")
plt.ylabel("Server Downtime (Hours)")
plt.show()
```

# Firewall Status Count

```python
plt.figure(figsize=(6, 4))
sns.countplot(x="Firewall Status", data=df, hue="Firewall Status", palette="Set2", legend=False)
plt.title("Firewall Status Distribution")
plt.xlabel("Firewall Status")
plt.ylabel("Count")
plt.show()
```

# Security Patches Analysis

```python
# Corrected code for Security Patches Applied countplot
plt.figure(figsize=(6, 4))
sns.countplot(x="Security Patches Applied", data=df, hue="Security Patches Applied", palette="Set1", legend=False)
plt.title("Security Patches Applied Status")
plt.xlabel("Security Patches Applied")
plt.ylabel("Count")
plt.show()
```

# Top 5 Installed Software

```python
plt.figure(figsize=(8, 4))
df["Installed Software"].value_counts().head(5).plot(kind="bar", color="purple")
plt.title("Top 5 Installed Software")
plt.xlabel("Software Name")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```
